## Common Methods

```python
import math

def f(n):
    if n == 0 or n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return f(n-1) + f(n-2)

def fib(n, memo={}):
    if n == 0 or n == 1:
        return 1
    elif n == 2:
        return 2
    try:
        return memo[n]
        pass
    except KeyError:
        f = fib(n-1) + fib(n-2)
        memo[n] = f
        return f

def is_prime(n):
    if n == 1:
        return False
    prime = True
    upper_lim = int(math.sqrt(n)) + 1
    for i in range(2, upper_lim):
        if n % i == 0:
            return False
    return prime

def factor(n):
    facts = set()
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            facts.add(int(i))
            facts.add(int(n/i))
    return list(sorted(facts))

def proper_factor(n):
    facts = set()
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            facts.add(int(i))
            facts.add(int(n/i))
    facts.remove(n)
    return list(sorted(facts))

def num_factors(n):
    facts = set()
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            facts.add(int(i))
            facts.add(int(n/i))
    return len(facts)

def is_palindrome(n):
    snum = str(n)
    if snum == snum[::-1]:
        return True
    return False
```

#### 5. Smallest Multiple


2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?


```python
def check_if_div_by_range(n, lo, hi):
    for i in range(lo, hi+1):
        if n%i != 0:
            return False
    return True

def smallest_multiple():
    for x in range(20, 300000000, 20):
        if check_if_div_by_range(x, 1, 20):
            return x
        
smallest_multiple()
```

#### 6. Sum Square Difference 

The sum of the squares of the first ten natural numbers is,
12 + 22 + ... + 102 = 385

The square of the sum of the first ten natural numbers is,
(1 + 2 + ... + 10)2 = 552 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.


```python
def sum_of_squares(n):
    sums = 0
    for x in range(n + 1):
        sums += x*x
    return sums

def square_of_sums(n):
    sums = 0
    for x in range(n + 1):
        sums += x
    return sums*sums

def sum_square_diff(n):
    return square_of_sums(n) - sum_of_squares(n)
    
sum_square_diff(100)
```

#### 9. Special Pythagorean Triplet

A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a<sup>2</sup> + b<sup>2</sup> = c<sup>2</sup>

For example, 3<sup>2</sup> + 4<sup>2</sup> = 9 + 16 = 25 = 52.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
def pythag_triples():
    lim = 1000
    for a in range(1, lim-2):
        for b in range(a+1, lim-1):
            for c in range(b+1, lim):
                if a**2 + b**2 == c**2:
                    # print(f'{a}^2 + {b}^2 == {c}^2')
                    if a + b + c == 1000:
                        return a*b*c
    #print(f'{i} {j} {k}')
pythag_triples()
```

#### 14 Longest Collatz sequence

The following iterative sequence is defined for the set of positive integers:

<pre>
n → n/2 (n is even)
n → 3n + 1 (n is odd)
</pre>

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.


```python
def collatz(start, debug=False):
    seq = []
    n = start
    seq.append(n)
    while n != 1:
        if debug: print(f'{n} -> ', end='')
        if n % 2 == 0:
            n = n//2
            seq.append(n)
        else:
            n = (3*n) + 1
            seq.append(n)
    if debug: print(n)
    return seq
    
def longest_collatz(start, max_num, debug=False):
    length = 0
    i = start
    num = i
    while i < max_num:
        new_length = len(collatz(i, debug))
        if new_length > length:
            length = new_length
            num = i
        i += 1
    return [num, length]

longest_collatz(1, 1000000)
        
```
