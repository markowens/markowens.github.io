### Daily Coding Problem: Problem #1 [Easy] (Complete)

Given a list of numbers and a number k, return whether any two numbers from the list add up to k.

For example, given [10, 15, 3, 7] and k of 17, return true since 10 + 7 is 17.

Bonus: Can you do this in one pass?

```python
def addup(nums, target):
    for i in range(len(nums)-1):
        for j in range(i+1, len(nums)):
            sm = nums[i] + nums[j]
            if sm == target:
                ans = str(nums[i]) + ' + ' + str(nums[j]) + ' == ' + str(target)
                return ans
    return False

def addup2(nums, target):
    sums = set([])
    for i in range(len(nums)-1):
        for j in range(i+1, len(nums)):
            sums.add(nums[i] + nums[j])
    if target in sums:
        return True
    else:
        return False

def aug28():
    nums = [23, 4, 2, 35, 12, 8, 92, 5, 3]
    target = 24
    print(addup(nums, target))
    print(addup(nums, 40))
    print(addup([10, 15, 3, 7], 17))
    print(addup2(nums, target))
    print(addup2(nums, 40))
    print(addup2([10, 15, 3, 7], 17))
        
aug28()
```

### Daily Coding Problem: Problem #2 [Hard] (Completed)

Given an array of integers, return a new array such that each element at index i of the new array is the product of all the numbers in the original array except the one at i.

For example, if our input was [1, 2, 3, 4, 5], the expected output would be [120, 60, 40, 30, 24]. If our input was [3, 2, 1], the expected output would be [2, 3, 6].

Follow-up: what if you can't use division?

```python
from functools import reduce

def prods(input):
    prod = reduce((lambda x, y: x * y), input)
    new_list = []
    for i in range(len(input)):
        new_list.append(prod//input[i])
    return new_list

def prod1(input):
    new_list = []   
    for i in range(len(input)):
        prod = 1
        for j in range(len(input)):
            if j != i:
                prod *= input[j]
        new_list.append(prod)
    return new_list
        

def aug29():
    inputs = [
        [1, 2, 3, 4, 5],
        [3, 2, 1],
        [4, 3, 6, 1, 2, 4, 3, 5]
    ]
    for i in inputs:
        output = prods(i)
        print(output)
        output = prod1(i)
        print(output)

aug29()
```

### Daily Coding Problem: Problem #4 [Hard]

This problem was asked by Stripe.

Given an array of integers, find the first missing positive integer in linear time and constant space. In other words, find the lowest positive integer that does not exist in the array. The array can contain duplicates and negative numbers as well.

For example, the input [3, 4, -1, 1] should give 2. The input [1, 2, 0] should give 3.

You can modify the input array in-place.

```python
def get_first_low_int(lst):
    lst.sort()
    print(lst)
    for i in range(len(lst)-1):
        # print(f"{i}: {lst[i]}")
        if lst[i] <= 0:
            continue
        if lst[i+1] - lst[i] != 1:
            return lst[i]+1
        
print(get_first_low_int([3, 4, -1, 1]))   
print(get_first_low_int([1, 2, 0]))
```
