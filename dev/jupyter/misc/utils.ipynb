## Parse Accumulo ITs for Failing Tests

The following snippet of code parses the error output from Accumulo IT tests looking for any tests containing failures or errors. It prints out the stats and file containing the failures

```python
import glob

ERROR_DIR = "/home/mark/dev/accumulo/test/target/failsafe-reports"

files = glob.glob(ERROR_DIR + "/*.txt")
output = []
for file in files:
    if "-output.txt" in file:
        continue
    with open(file) as f:
        for line in f:
            if "Errors: " in line:
                data = line.split()
                fname = data[-1] + ".txt"
                (tests, failures, errors, skipped) = (data[2][0], data[4][0], data[6][0], data[8][0])
                if int(failures) + int(errors)  > 0:
                    print(f'Failures: {failures},  Errors: {errors},  Skipped: {skipped} - {fname}')
                    output.append(fname)

for file in output:
    with open(ERROR_DIR + "/" + file) as f:
        for line in f:
            print(line.strip())
                    
                
```

## Compare Python Environments

Run <code>pip freeze > /tmp/requirements.txt</code> within your envrionment, renaming 'requirements.txt' to a name of your choice. Then compare the contents of the two files.

```python
GENERIC = '/tmp/generic'
GLOBAL = '/tmp/global'

def get_intersection(set1, set2):
    return set1.intersection(set2)

def print_set(sett, msg=None):
    if not msg == None:
        print(msg)
        print("-"*len(msg))
    for item in sett:
        print(item)

def get_difference(set1, set2):
    return set1 - set2

generic_set = set()
global_set = set()

with open(GENERIC) as generic:
    for line in generic:
        generic_set.add(line.strip().split('==')[0])
        #print(line.strip().split('==')[0])
        
with open(GLOBAL) as glbal:
    for line in glbal:
        global_set.add(line.strip().split('==')[0])   
        
intersection = get_intersection(generic_set, global_set)
print_set(intersection, "Modules in both generic and global envs")
    
diff1 = get_difference(generic_set, global_set)
print_set(diff1, "Modules in Generic that are NOT in Global")
    
diff1 = get_difference(global_set, generic_set)
print_set(diff1, "Modules in Global that are NOT in Generic")
```
