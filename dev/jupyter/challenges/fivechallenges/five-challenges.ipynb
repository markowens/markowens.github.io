### Question Marks

<p>
Have the function QuestionsMarks(str) take the str string parameter, which will contain single digit numbers, letters, and question marks, and check if there are exactly 3 question marks between every pair of two numbers that add up to 10. If so, then your program should return the string true, otherwise it should return the string false. If there aren't any two numbers that add up to 10 in the string, then your program should return false as well.
</p>
<p>
    For example: if str is <code>"arrb6???4xxbl5???eee5"</code> then your program should return true because there are exactly 3 question marks between 6 and 4, and 3 question marks between 5 and 5 at the end of the string.
</p>

#### Examples

<pre>
"arrb6???4xxbl5???eee5" => true
"acc?7??sss?3rr1??????5" => true
"5??aaaaaaaaaaaaaaaaaaa?5?5" => false
"9???1???9???1???9" => true
"aa6?9" => false
</pre>

```python
print("Question Marks")

str1 = "arrb6???4xxbl5???eee5"
str2 = "acc?7??sss?3rr1??????5"
str3 = "9???1???9???1???9"
str4 = "5??aaaaaaaaaaaaaaaaaaa?5?5"
str5 = "aa6?9"
str6 = "abc5???6???7???8abd"
str7 = "abc5??6???7???9abc"

"""
loop str:
ignore chars:
find nums:
"""
def count_marks(tmp):
    cnt = 0
    for c in tmp:
        if c.isalpha():
            continue
        if c == '?':
            cnt += 1
    return cnt

def check_marks(str, tup1, tup2):
    if int(tup1[0]) + int(tup2[0]) != 10:
        return True
    numqs = count_marks(str[tup1[1]+1:tup2[1]])
    if numqs != 3:
        return False
    return True

def check_for_tens(imap):
    for i in range(len(imap)-1):
        sum = int(imap[i][0]) + int(imap[i+1][0])
        if sum == 10:
            return True
    return False

def question_marks(str):
    result = True
    imap = [(str[i], i) for i in range(len(str)) if str[i].isdigit()]
    if not check_for_tens(imap):
        return False
    for i in range(len(imap)-1):
        if not check_marks(str, imap[i], imap[i+1]):
            return False
    return True
        
assert(question_marks(str1))
assert(question_marks(str2))
assert(question_marks(str3))
assert(not question_marks(str4))
assert(not question_marks(str5))
assert(not question_marks(str6))
assert(not question_marks(str7))
```
