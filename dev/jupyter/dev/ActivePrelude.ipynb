```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# v = (4/3)* math.pi * r^2
drain = 1.2
r = 3
v = (4/3) * math.pi * r**2
print(f"Total volume of sphere with radius {r} = {v}")


# Data for plotting
t = np.arange(0.0, 33.0, 1)

# s = 1 + np.sin(2 * np.pi * t)
s = 12 * math.pi - (drain * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='volume)',
       title='volume vs time')
ax.grid()

fig.savefig("test.png")
plt.show()
```
